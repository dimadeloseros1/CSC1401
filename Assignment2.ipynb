```python
# -*- coding: utf-8 -*-
"""
Created on Sat Mar 22 18:05:48 2025

@author: dimad
"""

# Assignment 1

def user_info():
    ''' This function is in charge of retrieving the information from
    the user before passing it to the next function.
    
    Returns the required input from the user.
    -------
    None.

    '''
    info = "";
    result = "";
    
    # The loop will run until the user inputs END word.
    while (info != "END"):    
        info = input("Please enter the letter (r, e, d, b, l, u) or END to submit the letters of your choice: ");
        
        if (info.upper() == "END"):     
              break;
        
        if (info != "r" and info != "e" and info != "d" and info != "b" and info != "l" and info != "u" and info != " " and info != "red" and info != "blue"):
            print("Please enter the correct letter from the previous choice of letters");
            continue;
        else:
            result += info;
            
    #We invoke the word_red_blue() function and provide it the argument of *result* which is the user input.
    display_word(result);
    return result;


def letter_r():
    '''
    Return a string which will print the alphabet pattern 'r', shown below.
    
    The letter fits in a 6x6 grid
    
        *  * 
        * *  
        *   
        *   
        *   

    '''
    result = "";
    for row in range(6):
        for column in range(6):
            if ((column == 3 and row != 0 or (column == 4 and row == 2) or (column == 5 and row == 1))):
                result += "* ";
            else:
                result += "  ";
        result += "\n";
    return result;
letter_r();

def letter_e():
    '''
    Return a string which will print the alphabet pattern 'e'
    
    The letter fits in a 6x6 grid
        *     
      *   *   
      * * *   
      *       
        * * 
    
    '''
    result = ""
    for row in range(6):
        for column in range(6):
            if ((row == 1 and column == 3) or (row == 2 and (column == 2 or column == 4)) or
                (row == 3 and (column == 2 or column == 3 or column == 4)) or (row == 4 and column == 2) or
                (row == 5 and column == 3) or (row == 5 and column == 4)): 
                result += "* ";
            else:
                result += "  ";
        result += "\n"
    
    return result;

letter_e();



def letter_d():
    """ Return a string which will print the alphabet pattern 'd', shown below.
    
		The letter fits in a 6x6 grid
        
          *  
          *  
        * *  
      *   *  
        * *           
    """    

    result_str = "";
    for row in range(6):
        for column in range(6):
            result_str += " "
            if ((column == 2 and row == 4) or (column == 3 and row == 3) or (column == 3 and row == 5) or (column == 4 and row != 0)):
                result_str += "*";
            else:
                result_str += " ";
        result_str += "\n";
    # print(result_str);
    return result_str;

letter_d();


def letter_b():
    """ Return a string which will print the alphabet pattern 'b', shown below.
    
		The letter fits in a 6x6 grid
        
          *      
          *      
          * *    
          *   *  
          * *              
    """    

    result_str = "";
    for row in range(6):
        for column in range(6):
            result_str += " "
            if ((column == 4 and row == 4) or (column == 3 and row == 3) or (column == 3 and row == 5) or (column == 2 and row != 0)):
                result_str += "*";
            else:
                result_str += " ";
        result_str += "\n";
    # print(result_str);
    return result_str;

letter_b();


def letter_l():
    """ Return a string which will print the alphabet pattern 'l', shown below.
        
		The letter fits in a 6x6 grid
        
          *    
          *    
          *    
          *    
          * *         
    """    

    result_str = "";
    for row in range(6):
        for column in range(6):
            result_str += " "
            if ((column == 3 and row != 0) or (column == 4 and row == 5)):
                result_str += "*";
            else:
                result_str += " ";
        result_str += "\n";
    # print(result_str);
    return result_str;

letter_l();



def letter_u():
    """ Return a string which will print the alphabet pattern 'u', shown below.
       
		The letter fits in a 6x6 grid
        
          *   *  
          *   *  
          *   *  
          *   *  
            *               
    """    

    result_str = "";
    for row in range(6):
        for column in range(6):
            result_str += " ";
            if ((column == 2 or column == 4) and (row != 5 and row != 0) or (column == 3 and row == 5)):
                result_str += "*";
            else:
                result_str += " ";
        result_str += "\n";
    # print(result_str);
    return result_str;

letter_u();

def display_word(result):
    '''
    

    Parameters
    ----------
    result : TYPE: String
        We will use the parameter of this function to pass the user input
        This function is in-charge of orchestrating each of the letter to be printed as per the assignment requirement

    Returns the final result of the function as a String
    -------

    '''
    # We create an array which will alow us to store multiple letters as well as single at once
    # We use the .split('\n) function to convert the letter pattern into a list of rows
    # Each letter created will be appended to the empty array that we've created on line 217
    letter_arr = []
    for letter in result:
        if (letter == "r"):
           letter_arr.append(letter_r().split('\n'))
        elif (letter == "e"):
            letter_arr.append(letter_e().split('\n'))
        elif (letter == "d"):
            letter_arr.append(letter_d().split('\n'))
        elif (letter == "b"):
            letter_arr.append(letter_b().split('\n'))
        elif (letter == "l"):
            letter_arr.append(letter_l().split('\n'))
        elif (letter == "u"):
            letter_arr.append(letter_u().split('\n'))
   
    # Print letters side by side, in this case *red* and *blue* as well as single letters
    final_result = ""
    for row in range(6):  
        for arr in letter_arr:
            final_result += arr[row];
        final_result += "\n";
   
    print(final_result);
    return final_result;

def main():
    '''
        This is the main function and it is in-charge of orchestrating the workflow of the application
        by invoking the user_info() function
    '''
    user_info();

if __name__ == "__main__":
    main();
```
